```python
                                                 REMOVE UNWANTED CHARACTER
```

```python
import pandas as pd

df = pd.DataFrame({
    'Product': [
        'Laptop#Dell!', 
        'Smartphone@Samsung', 
        'Keyboard (Wireless)', 
        'HDMI Cable***', 
        'USB-C Adapter%20'
    ]
})

df['Cleaned_Product'] = (df['Product']
    .str.replace(r'[^\w\s]', '', regex=True)  
    .str.strip()                              
    .str.lower()                             
)

print(df)
```

```python
                                               REMOVE COLUMN
```

```python
import pandas as pd

df = pd.DataFrame({
    'Country': ['PK', 'IN', 'US'],
    'Population': [240, 1400, 330],
    'Continent': ['Asia', 'Asia', 'North America'],
    'Temp_Info': [None, 45, None]
})

print("Original:")
print(df)


df2 = df.drop('Temp_Info', axis=1)
print("\nAfter drop('Temp_Info'):\n", df2)

del df2['Continent']
print("\nAfter del df2['Continent']:\n", df2)

pop_col = df2.pop('Population')
print("\nAfter pop(), remaining df2:\n", df2)
print("\nPopped column:", pop_col)
```

```python
                                         FILTER DATA
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Usman', 'Khan', 'Zara'],
    'Age':  [23,    30,     22,     35,     28],
    'City': ['Lahore', 'Karachi', 'Islamabad', 'Lahore', 'Karachi'],
    'Salary': [50000, 60000, 45000, 75000, 62000]
})


filter1 = df[df['Age'] > 28]

filter2 = df[(df['Age'] > 25) & (df['Salary'] > 60000)]

filter3 = df[df['City'].isin(['Lahore','Islamabad'])]

filter4 = df.query('Age > 25 and City == "Lahore"')

print(filter1)
print(filter2)
print(filter3)
print(filter4)
```

```python
                                          DROP COLUMN
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Usman', 'Khan'],
    'Age': [25, 30, 22, 35],
    'City': ['Lahore', 'Karachi', 'Islamabad', 'Lahore'],
    'Salary': [50000, 60000, 45000, 75000]
})

print("Original DataFrame:")
print(df)

df = df.drop('Salary', axis=1)
print("\nAfter dropping 'Salary':")
print(df)

df = df.drop(['City'], axis=1)
print("\nAfter dropping 'City':")
print(df)
```
